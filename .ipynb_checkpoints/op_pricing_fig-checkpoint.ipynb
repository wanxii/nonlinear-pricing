```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import SubplotSpec
import seaborn as sns
import numpy as np
import pandas as pd
from math import ceil
```

```python
def generate_utility(v1, v2, q1, q2, n):
    length = max(q1 + 1 + n, q2 + 1 + n)
    array_q = np.arange(length)
    array_u1 = np.concatenate((v1 * array_q[:q1 + 1], v1 * q1 * np.ones(length - q1 - 1)))
    array_u2 = np.concatenate((v2 * array_q[:q2 + 1], v2 * q2 * np.ones(length - q2 - 1)))
    return array_q, array_u1, array_u2
```

### Positive Relationship between $v$ and $\bar{q}$

```python
vl, vh, ql, qh = 1, 4, 3, 5
v1, v2, q1, q2 = vh, vl, qh, ql
q, u1, u2 = generate_utility(v1, v2, q1, q2, 3)
beta0 = q2 * (v2 - v1)

plt.plot(q, u1, label = "Consumer1")
plt.plot(q, u2, label = "Consumer2")
plt.plot(q, v1 * q,
         linestyle="--",
         label = "Pricing1",
         color = "black")
plt.plot(q, np.concatenate((u2[:q2], beta0 + v1 * q[q2:])),
         linestyle="-.",
         label = "Pricing2",
         color = "black")
# plt.plot(q, np.concatenate((u2[:q2 - 1], beta0 + v1 * (q[q2 - 1:] + 1))),
#          linestyle = "--",
#          color = "black")

for x, ym in {# q2 - 2: u1[q2 - 2],
              q2: u1[q2],
              q1: u1[q1]}.items():
    plt.vlines(x = x, ymin = 0, ymax = ym,
               linestyle = ":",
               colors = "gray")
plt.legend();

plt.savefig("fig/op_pricing.svg", dpi=400)
```

### Negative Relationship between $v$ and $\bar{q}$

CASE 1: $\frac{v_h}{v_l} < \frac{\bar{q}_h}{\bar{q}_l}$

```python
vl, vh, ql, qh = 1, 2, 2, 5
v1, v2, q1, q2 = vh, vl, ql, qh
q, u1, u2 = generate_utility(v1, v2, q1, q2, 3)

beta1 = (v2 * q2 - v1 * q1) / (q2 - q1)
beta0 = q1 * q2 * (v1 - v2) / (q2 - q1)

plt.plot(q, u1, label = "Consumer1")
plt.plot(q, u2, label = "Consumer2")
plt.plot(q, np.concatenate((u1[:ql], beta0 + beta1 * q[ql:])),
         linestyle="--",
         label = "Optimal Pricing",
         color = "black")
plt.legend();

plt.savefig("fig/op_pricing_case1.svg", dpi=400)
```

CASE 2: $\frac{v_h}{v_l} \geq \frac{\bar{q}_h}{\bar{q}_l}$

(a) $(v_h - c_2)q_l \leq (v_l - c_2)q_h + v_h\lfloor \frac{v_l q_h}{v_h} \rfloor - c_2 q_l$

```python
vl, vh, ql, qh = 1, 2, 3, 5
v1, v2, q1, q2 = vh, vl, ql, qh
q, u1, u2 = generate_utility(v1, v2, q1, q2, 3)
tp = v2 * q2 / v1
q_p = sorted(np.append(q, tp))

plt.plot(q, u1, label = "Consumer1")
plt.plot(q, u2, label = "Consumer2")
plt.plot(q_p, np.concatenate((u1[:ceil(tp)], v2 * q2 * np.ones(len(u1) - ceil(tp) + 1))),
         linestyle="--",
         label = "Optimal Pricing",
         color = "black")
plt.legend();

plt.savefig("fig/op_pricing_case2a.svg", dpi=400)
```

(b)  $(v_h - c_2)q_l > (v_l - c_2)q_h + v_h\lfloor \frac{v_l q_h}{v_h} \rfloor - c_2 q_l$

```python
vl, vh, ql, qh = 0.5, 2, 3, 5
v1, v2, q1, q2 = vh, vl, ql, qh
q, u1, u2 = generate_utility(v1, v2, q1, q2, 3)

plt.plot(q, u1, label = "Consumer1")
plt.plot(q, u2, label = "Consumer2")
plt.plot(q, v1 * q,
         linestyle="--",
         label = "Optimal Pricing",
         color = "black")
plt.legend();

plt.savefig("fig/op_pricing_case2b.svg", dpi=400)
```

### Special Example

#### 2 Subplots

```python
# Define v, q values
vl, vh, ql, qh = 1200, 2000, 10, 20
v1, v2, q1, q2 = vh, vl, qh, ql
v1_1, v2_1, q1_1, q2_1 = vl, vh, qh, ql

# Calcuate utilities
q, u1, u2 = generate_utility(v1, v2, q1, q2, 5)
q_1, u1_1, u2_1 = generate_utility(v1_1, v2_1, q1_1, q2_1, 5)

# Create subplots
figure, axis = plt.subplots(2, 1, figsize = (8, 12))
ax1, ax2 = axis
plt.subplots_adjust(hspace = 0.5)

# Positive Correlation
beta0 = q2 * (v2 - v1)

## Payment
ax1.plot(q, u1, label = "Consumer1")
ax1.plot(q, u2, label = "Consumer2")
## Optimal Pricing
ax1.plot(q, np.concatenate((u2[:q2], q2 * (v2 - v1) + v1 * q[q2:])),
         linestyle="--",
         linewidth = 3,
         label = "Optimal Pricing",
         color = "black")
## Lines and dots
for x, ym in {q1: u1[q1] + beta0, q2: u2[q2]}.items():
    ax1.vlines(x = x, ymin = 0, ymax = ym,
               linestyle = ":",
               colors = "gray")
    ax1.hlines(y = ym, xmin = 0, xmax = x,
               linestyle = ":",
               colors = "gray")
    ax1.plot([x, 0, x], [ym, ym, 0], "ko", ms = 7, clip_on = False)
ax1.vlines(x = q1, ymin = u1[q1] + beta0, ymax = u1[q1],
           linestyle = ":",
           colors = "gray")
## Add tittle
ax1.set_title(r"Positive Correlation Between $\bar{q}$ and $v$, $\bar{q}_1$ = %s, $v_1$ = %s, $\bar{q}_2$ = %s, $v_2$ = %s" % (q1, v1, q2, v2),
              fontsize = 18,
              pad = 35)
ax1.legend()

# Negative Correlation
beta1 = (v2_1 * q2_1 - v1_1 * q1_1) / (q2_1 - q1_1)
beta0 = q1_1 * q2_1 * (v1_1 - v2_1) / (q2_1 - q1_1)

## Payment
ax2.plot(q_1, u1_1, label = "Consumer1")
ax2.plot(q_1, u2_1, label = "Consumer2")
## Optimal Pricing
plt.plot(q, np.concatenate((u1[:ql], beta0 + beta1 * q[ql:])),
         linestyle="--",
         linewidth = 3,
         label = "Optimal Pricing",
         color = "black")
## Lines and dots
for x, ym in {q1_1: u1_1[q1_1], q2_1: u2_1[q2_1]}.items():
    ax2.vlines(x = x, ymin = 0, ymax = ym,
               linestyle = ":",
               colors = "gray")
    ax2.hlines(y = ym, xmin = -1, xmax = x,
               linestyle = ":",
               colors = "gray")
    ax2.plot([x, 0, x], [ym, ym, 0], "ko", ms = 7, clip_on = False)
## Add tittle
ax2.set_title(r"Negative Correlation Between $\bar{q}$ and $v$, $\bar{q}_1$ = %s, $v_1$ = %s, $\bar{q}_2$ = %s, $v_2$ = %s" % (q1_1, v1_1, q2_1, v2_1),
              fontsize = 18,
              pad = 35)
ax2.set_xlabel("Value",
               size = 15,
               labelpad = 20)

# Format axes
for ax in axis:
    ax.set_xlim([0, max(q)])
    ax.set_ylim([0, 45000])
    ax.set_ylabel("Utility / Payment",
                  size = 15,
                  labelpad = 20)
    sns.despine(ax = ax, offset = 0)
    ax.tick_params(labelsize = 12)

plt.savefig("fig/comparison2.svg", dpi=400)
```

#### 4 Subplots

```python
# Function for adding common subtitle per row
def create_subtitle(fig: plt.Figure, grid: SubplotSpec, title: str):
    row = fig.add_subplot(grid)
    # the '\n' is important
    row.set_title(f'{title}\n',
                  fontsize = 18,
                  pad = 20)
    # hide subplot
    row.set_frame_on(False)
    row.axis('off')
```

```python
# Define v, q values
vl, vh, ql, qh = 1200, 2000, 10, 20
v1, v2, q1, q2 = vh, vl, qh, ql
v1_1, v2_1, q1_1, q2_1 = vl, vh, qh, ql

# Calcuate utilities
q, u1, u2 = generate_utility(v1, v2, q1, q2, 5)
q_1, u1_1, u2_1 = generate_utility(v1_1, v2_1, q1_1, q2_1, 5)

# Layout subplots
widths = [1, 2]
heights = [1, 1]
gs_kw = dict(width_ratios=widths, height_ratios=heights)
figure, axes = plt.subplots(ncols=2, nrows=2,
                            gridspec_kw=gs_kw,
                            figsize = (10, 10))
ax11, ax12, ax21, ax22 = axes[0, 0], axes[0, 1], axes[1, 0], axes[1, 1]
plt.subplots_adjust(wspace = 0.5, hspace = 0.6)


# Positive correlation
beta0 = q2 * (v2 - v1)

## Payment
ax12.plot(q, u1, label = "Consumer1")
ax12.plot(q, u2, label = "Consumer2")
## Optimal Pricing
ax12.plot(q, np.concatenate((u2[:q2], q2 * (v2 - v1) + v1 * q[q2:])),
         linestyle="--",
         linewidth = 3,
         label = "Optimal Pricing",
         color = "black")
## Lines and dots
for x, ym in {q1: u1[q1] + beta0, q2: u2[q2]}.items():
    ax12.vlines(x = x, ymin = 0, ymax = ym,
               linestyle = ":",
               colors = "gray")
    ax12.hlines(y = ym, xmin = 0, xmax = x,
               linestyle = ":",
               colors = "gray")
    ax12.plot([x, 0, x], [ym, ym, 0], "ko", ms = 7, clip_on = False)
ax12.vlines(x = q1, ymin = u1[q1] + beta0, ymax = u1[q1],
           linestyle = ":",
           colors = "gray")
ax12.legend()

# Negative correlation
beta1 = (v2_1 * q2_1 - v1_1 * q1_1) / (q2_1 - q1_1)
beta0 = q1_1 * q2_1 * (v1_1 - v2_1) / (q2_1 - q1_1)

## Payment
ax22.plot(q_1, u1_1, label = "Consumer1")
ax22.plot(q_1, u2_1, label = "Consumer2")
## Optimal Pricing
plt.plot(q, np.concatenate((u1[:ql], beta0 + beta1 * q[ql:])),
         linestyle="--",
         linewidth = 3,
         label = "Optimal Pricing",
         color = "black")
## Lines and dots
for x, ym in {q1_1: u1_1[q1_1], q2_1: u2_1[q2_1]}.items():
    ax22.vlines(x = x, ymin = 0, ymax = ym,
               linestyle = ":",
               colors = "gray")
    ax22.hlines(y = ym, xmin = -1, xmax = x,
               linestyle = ":",
               colors = "gray")
    ax22.plot([x, 0, x], [ym, ym, 0], "ko", ms = 7, clip_on = False)

# Scatter plots
for Q, (i, V) in {q1: (1, v1), q2: (2, v2)}.items():
    ax11.scatter(Q, V,
                 linewidths = 3,
                 label = "Consumer{}".format(i))

ax11.legend()

for Q, V in {q1_1: v1_1, q2_1: v2_1}.items():
    ax21.scatter(Q, V,
                 linewidths = 3)

# Format axis
for ax in [ax12, ax22]:
    ax.set_xlim([0, max(q)])
    ax.set_ylim([0, 45000])
    ax.set_ylabel("Utility / Payment",
                  size = 15,
                  labelpad = 20)
    sns.despine(ax = ax, offset = 0)
    ax.tick_params(labelsize = 12)

for ax in [ax11, ax21]:
    ax.set_xticks(np.arange(0, max(q) + 1, 5))
    ax.set_ylim([0, 1.5 * vh])
    ax.set_ylabel("Value",
                  size = 15,
                  labelpad = 20)
    sns.despine(ax = ax, offset = 0)
    ax.tick_params(labelsize = 12)

figure.supxlabel("Quantity",
                 size = 15)

# Add tittles
grid = plt.GridSpec(2, 2)
create_subtitle(figure, grid[0, ::],
                r"Positive Correlation Between $\bar{q}$ and $v$, $\bar{q}_1$ = %s, $v_1$ = %s, $\bar{q}_2$ = %s, $v_2$ = %s" % (q1, v1, q2, v2))
create_subtitle(figure, grid[1, ::],
                r"Negative Correlation Between $\bar{q}$ and $v$, $\bar{q}_1$ = %s, $v_1$ = %s, $\bar{q}_2$ = %s, $v_2$ = %s" % (q1_1, v1_1, q2_1, v2_1))

plt.savefig("fig/comparison4.svg", dpi=400)
```
